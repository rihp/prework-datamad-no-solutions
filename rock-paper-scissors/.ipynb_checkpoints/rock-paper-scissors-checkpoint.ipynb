```python
# Import the choice function of the random module
# https://stackoverflow.com/questions/306400/how-to-randomly-select-an-item-from-a-list

from random import choice 

# Assign to a list the 3 possible options: 'rock', 'paper' or 'scissors'.
options = ['rock', 'paper', 'scissors']
rock = options[0]
paper = options[1]
scissors = options[2]

# Assign a variable to the maximum number of games: 1, 3, 5, etc ...
max_games = 5

# Assign a variable to the number of games a player must win to win.
# Preferably the value will be based on the number of maximum games
g2w = 3

# Define a function that randomly returns one of the 3 options.
# This will correspond to the play of the machine. Totally random.
def machine_play():
    machine_play = choice(options)
    print('machine chose: ', machine_play)
    return machine_play 

    
# Define a function that asks your choice: 'rock', 'paper' or 'scissors'
# you should only allow one of the 3 options. This is defensive programming.
# If it is not rock, paper or scissors keep asking until it is.

def user_play():
    valid_move = False
    while valid_move is False:
        user_play = input('Choose a move to play: rock, paper, or scissors?  -')
        if user_play in options:
            valid_move = True
        elif user_play not in options:
            print('invalid move.')
            valid_move = False
    print('user chose: ', user_play)
    return user_play

# Define a function that resolves a combat.
# Returns 0 if there is a tie, 1 if the machine wins, 2 if the human player wins

outcome = 'no games played'

def game():
    global outcome
    print('one, two, three!')

    if user_play_value == rock:
        if machine_play_value == rock:
            outcome = 'tie'
        elif machine_play_value == paper:
            outcome = 'machine'    
        elif machine_play_value == scissors:
            outcome = 'user'
        else:
            return print('error no 001, game function failed')

    elif user_play_value == paper:
        if machine_play_value == rock:
            outcome = 'user'
        elif machine_play_value == paper:
            outcome = 'tie'        
        elif machine_play_value == scissors:
            outcome = 'machine'
        else:
            return print('error no 001, game function failed')
    
    elif user_play_value == scissors:
        if machine_play_value == rock:
            outcome = 'machine'
        elif machine_play_value == paper:
            outcome = 'user'        
        elif machine_play_value == scissors:
            outcome = 'tie'
        else:
            return print('error no 001, game function failed')
    return outcome

```
